```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.gridspec as gridspec

# Generaración de DataFrame
np.random.seed(42)
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
         "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]
ventas = np.random.randint(2000, 5000, 12)
gastos = np.random.randint(1000, 4000, 12)

df = pd.DataFrame({
    'Mes': meses,
    'Ventas': ventas,
    'Gastos': gastos
})

# Crear figura con GridSpec
fig = plt.figure(figsize=(20, 15))
gs = gridspec.GridSpec(3, 3, figure=fig)

# Primer gráfico (barras)
ax1 = fig.add_subplot(gs[0, :2])
ax1.bar(df['Mes'], df['Ventas'], color="blue")
ax1.set_title('Ventas por Mes')
ax1.set_ylabel('Ventas')
ax1.tick_params(axis='x', rotation=60, labelsize=9)

# Segundo gráfico (línea)
ax2 = fig.add_subplot(gs[0, 2])
ax2.plot(df['Mes'], df['Gastos'], marker='o', color='Red')
ax2.set_title('Gastos por Mes')
ax2.set_ylabel('Gastos')
ax2.tick_params(axis='x', rotation=60, labelsize=9)

# Tercer gráfico (dispersión)
ax3 = fig.add_subplot(gs[1:, :])
ax3.scatter(df['Ventas'], df['Gastos'], color='green')
ax3.set_title('Relación Ventas vs Gastos')
ax3.set_xlabel('Ventas')
ax3.set_ylabel('Gastos')

# Ajuste de espacios
plt.tight_layout()

plt.show()
```
