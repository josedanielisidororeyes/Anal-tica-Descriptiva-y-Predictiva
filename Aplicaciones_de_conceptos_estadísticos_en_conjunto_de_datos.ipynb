# 1) Operaciones básicas

```python
# 1.  Crear un array de 10 elementos con valores entre 1 y 10
import numpy as np
array_10 =  np.arange(1, 11)
print("---Impresión del Arreglo---")
print(array_10)
print("\n")

# 2. Calculo de media, mediana, y desviación estándar de los valores
media  =  np.mean(array_10)
mediana  =  np.median(array_10)
desviación_estandar =  np.std(array_10)

print(f"La media del arreglo es: {media}")
print(f"La mediana del arreglo es: {mediana}")
print(f"La desviación estándar del arreglo es: {desviación_estandar}")
print("\n")
#3. Multiplicación del arreglo por 2
print("---Impresión del Arreglo multiplicado por 2 ---")
print(array_10 * 2)
```

#2) Operaciones con DataFrames

```python
#1. Creación de DataFrame con 3 columnas y 5 números enteros aleatorios
import pandas as pd
np.random.seed(5)
datos_aleatorios  =  np.random.randint(1, 20, size =  (5, 3))
df =  pd.DataFrame(datos_aleatorios)
print("\n---Impresión de DataFrame---")
print(df)
print("\n")

#2. Renombre de columnas
df.columns =  ["Col1", "Col2", "Col3"]
print("\n---Impresión de DataFrame con columnas renombradas")
print(df)
#3. Nueva columna Col1 + Col2
df["Col1_mas_Col2"] =  df['Col1'] +  df['Col2']
print("\n")

#4. Eliminación de filas donde Col3 es menor a 10
print("\n---Impresión de Filas donde Col3 es menor a 10---")
df[df["Col3"] < 10]
```

#3) Estadística Descriptiva

```python
np.random.seed(5)
# 1. Creación de DataFrame
datos_100 =  np.random.randint(1, 51, size = (100, 1))
df =  pd.DataFrame(datos_100)
df.columns =  ["Col1"]

#2. Calculo de media, mediana, y moda
media  =  df['Col1'].mean()
mediana =  df['Col1'].median()
moda  =  df['Col1'].mode()[0]

print(f"La media del dataframe es: {media}")
print(f"La mediana del dataframe es: {mediana}")
print(f"La moda del dataframe es: {moda}")
print("\n")

#3. Calculo del valor mínimo y máximo

min =  df['Col1'].min()
max =  df['Col1'].max()

print(f"El valor mínimo del dataframe es: {min}")
print(f"El valor máximo del dataframe es: {max}")
```

#4. Análisis de Series Temporales

```python
# 1. Creación de ccolumna con fecha
fechas  = pd.date_range(start = "2025-01-01",  periods  =  100, freq = 'D')

# 2. Creación de 100 valores aleatorios
valores  = np.random.rand(100)

df =  pd.DataFrame(
    {'Fecha': fechas,
     'Valores': valores}
)

# Cálculo de la media movil de 5 días
df["Media_Movil_5"] =  df['Valores'].rolling(window  =  5).mean()
df.dropna(inplace  =  True)

print("\nImpresión de Primeras 5 Filas de la Media Movil de 5 días")
df["Media_Movil_5"].head()
```

# 10) Gráfico de Línea con Media Movil

```python
np.random.seed(5)

fechas =  pd.date_range(start  =  "2025-01-01", periods  =  100, freq =  "D")
valores =   np.random.rand(100)

df = pd.DataFrame(
    {"Fechas": fechas,
     'valores': valores}
)

df["Media_Movil_10"] =  df["valores"].rolling(window  =  10).mean()
df.dropna(inplace  =  True)

df.set_index("Fechas", inplace  =  True)
ax = df.plot(figsize=(9, 6), title="Serie Aleatoria  y Media Móvil de 10 Días")
ax.set_xlabel("")
ax.set_ylabel("Valor")

plt.show()
```

# 11) Datos Temporales con Análisis Estadístico


```python
np.random.seed(5)
fechas =  pd.date_range(start  =  "2025-01-01", periods  =  200, freq =  "D")
valores =   np.random.rand(200)

serie  =  pd.Series(valores, index  =  fechas)
media  =  serie.mean()
desv_std =  serie.std()

plt.figure(figsize =  (15, 8))
plt.plot(serie, label = "Serie Temporal", color  = "blue")
plt.axhline(media, color  = "red", linestyle  =  "--", label = "Media")
plt.axhline(media  +  desv_std, color  =  "green", linestyle  = "--", label  = "Media + 1σ")
plt.axhline(media - desv_std, color   =  "green", linestyle =  "--", label = "Media  - 1σ")

plt.title("Serie Semporal con bandas de confianza")
plt.xlabel("Fecha")
plt.ylabel("Valor")
plt.legend()
```
