## Ejercicio 01: Tipo de Datos Correcto

*   Carga el conjunto de datos airdata.csv y verifica que cada columna tenga el tipo de datos correcto

*   Si una columna de tipo temporal no está en el tipo de correcto realiza la conversión.

*   Usa dtypes para mostrar que cada columna, principalmente la que contiene datos temporales, está en el tipo de datos correcto
*   Imprime las primeras tres líneas




```python
import pandas as pd
import datetime as dt

# Carga del conjunto de datos
df =  pd.read_csv('/content/drive/MyDrive/Programación Analítica Predictiva/Airdata.csv')

# Verificación de tipo de datos de las columnas
print("---Tipo de Datos por Columna Antes de la Limpieza---\n")
df.info()
```

Podemos observar que la columna DateTime está almacenada como texto y no como fecha. Por lo tanto, no podemos analizar los datos como una serie temporal.

```python
# Tranformación de "DateTime" a tipo fecha
df['DateTime'] =  pd.to_datetime(df['DateTime'])

# Comprobación de tipo de dato por columna
print("---Tipos de Dato por Columna Despúes de la Limpieza---\n")
print(df.dtypes)

# Impresión de las primera 3 filas
df.head(2)
```

Después de la corrección, todas las columnas tienen el tipo de dato correcto. Ahora podemos trabajar con DateTime como fecha. Aunque la tarea no lo solicita, es muy recomendable fijar esta columna como indice para poder utilizar todas las funcionalidades de series temporales.

#Ejercicio 02: Extraer componentes de fecha y tiempo

*   De la columna 'DateTime' extrae el año, mes, día y hora. Crea una nueva columna para cada componente que extraes

*   Imprime las primeras tres líneas del dataframe que ahora deberá tener 4 nuevas columnas.


```python
# Extracción del año
df['Año'] =  df['DateTime'].dt.year

# Extracción del mes
df['Mes'] =  df['DateTime'].dt.month

# Extracción del día
df['Día'] =  df['DateTime'].dt.day

# Extracción de la hora
df['Hora'] =  df['DateTime'].dt.hour

# Impresión de las primera 3 filas
df.head(3)
```

#Ejercicio 03: Aritmética - Suma de Intervalos (timedelta)



*    Suma 10 días, 7 horas y 15 minutos a cada elemento de la columna 'DateTime'. Guarda el resultado en una nueva columna y agrega al dataframe.


*   Imprime la columna 'DateTime' y la columna que resulta de sumar 10 días, 7 horas y 15 minutos.


```python
# Suma de intervalos a DateTime
df['DateTime_Delta'] = df['DateTime'] + dt.timedelta(days = 10, hours = 7, minutes =  15)

# Impresión de DateTime y DateTime con suma de intervalos
print(df[['DateTime', 'DateTime_Delta']])
```

#Ejercicio 04: Timestamp



*   Convierte la columna 'DateTime' la cual tiene un formato de fecha legible a a timestamp Unix. Guarda el resultado en una nueva columna
*   Imprime las primeras tres filas de  la columna 'DateTime' y la columna que representa dicha fecha en timestamp Unix.


```python
# Conversión de DateTime a formato timestamp Unix
df['DateTime_Unix'] =  df['DateTime'].apply(lambda x: int(x.timestamp()))

# Impresión de las primera 3 filas de la columna original y transformada
print(df[['DateTime', 'DateTime_Unix']].head(3))
```

Como se puede observar la columna transformada DateTime_Unix muestra la cantidad de segundos transcurridos desde la época UNIX. A modo de ilustración, se muestra una solución alternativa usando pandas, la cual trabaja de manera vectorizada en lugar de fila a fila.

```python
# Se divide entre 10**9 para obtener los segundos,  ya que pandas regresa el tiempo en nanosegundos
df['DateTime_Unix_Pandas'] =   df['DateTime'].astype('int64') // 10**9
print(df[['DateTime', 'DateTime_Unix', 'DateTime_Unix_Pandas']].head(3))
```
