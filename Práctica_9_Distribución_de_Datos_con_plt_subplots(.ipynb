```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creación de DataFrame 1000 x 4
Normal  =  np.random.randn(1000)
Uniforme = np.random.uniform(-5, 5, 1000)
Exponencial = np.random.exponential(1, 1000)
Poisson  =  np.random.poisson(5, 1000)

df =  pd.DataFrame(
    {'Normal': Normal,
     'Uniforme': Uniforme,
     'Exponencial': Exponencial,
     'Poisson': Poisson}
)

# Creación de matriz de subplots 2x2
fig, ax =  plt.subplots(2, 2, figsize =  (10, 6))

# Gráfico 1 Distribución Normal
ax[0, 0].hist(df['Normal'], bins  =  30)
ax[0, 0].set_title('Distribución Normal')
ax[0, 0].set_ylabel('Frencuencia')

# Gráfico 2 Distribución Uniforme
ax[0, 1].hist(df['Uniforme'], bins  =  30)
ax[0, 1].set_title('Distribución Uniforme')
ax[0, 1].set_ylabel('Frencuencia')

# Grafico 3 Disgtribucion Exponencial
ax[1, 0].hist(df['Exponencial'], bins  =  30)
ax[1, 0].set_title('Distribución Exponencial')
ax[1, 0].set_ylabel('Frencuencia')

# Gráfico 4 Distribucion Poisson
ax[1, 1].hist(df['Poisson'], bins =  30)
ax[1, 1].set_title('Distribución Poisson')
ax[1, 1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
