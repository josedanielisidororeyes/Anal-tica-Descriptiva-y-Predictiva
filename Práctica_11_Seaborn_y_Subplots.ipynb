```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("penguins")

# Figura con 2 filas y 2 columnas
fig, ax =  plt.subplots(2, 2, figsize =  (10, 15))

# Histograma de las longitudes de las aletas
sns.histplot(df['flipper_length_mm'], kde  =  True, ax =  ax[0, 0])
ax[0, 0].set_title('Distribucin de las longitudes de aletas')
ax[0, 0].set_ylabel('Frencuencia')
ax[0, 0].set_xlabel('Largo de la Aleta (mm)')

# Boxplot Masa Corporal por Especie
sns.boxplot(x = 'species', y = 'body_mass_g', data =  df, palette =  'coolwarm', ax =  ax[0, 1])
ax[0, 1].set_title('Distribucin de Masa Corporal por Especie')
ax[0, 1].set_ylabel('Masa Corporal')
ax[0, 1].set_xlabel('Especie')

# Dispersin entre masa corporal y largo de aletas
sns.scatterplot(x = 'flipper_length_mm', y = 'body_mass_g', data =  df,  ax =  ax[1, 0], hue = 'species')
ax[1, 0].set_title('Relacin masa corporal y largo de aletas por especie')
ax[1, 0].set_xlabel('Largo de la Aleta')
ax[1, 0].set_ylabel('Masa Corporal')
ax[1, 0].set_xlabel('Especie')

# Violin Plot Masa Corporal por Especie
sns.violinplot(x = 'species', y = 'body_mass_g', data =  df, ax =  ax[1, 1], palette = 'coolwarm')
ax[1, 1].set_title('Distribucin de Masa Corporal  por Especie')
ax[1, 1].set_xlabel('Especie')

plt.tight_layout()
```
