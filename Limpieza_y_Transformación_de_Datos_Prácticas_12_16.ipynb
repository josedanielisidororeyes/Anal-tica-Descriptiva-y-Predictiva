# Practica 11: Carga de Datos

```python
import pandas as pd
# Carga de Datos
df = pd.read_csv('/content/drive/MyDrive/Programación Analítica Predictiva/miaad-nyc-r-s.csv')

# Eliminación de ultima fila
df =  df.iloc[:-1]

# Impresión de filas y columnas
print(f"El número de filas es: {df.shape[0]}\nEl número de columnas es: {df.shape[1]}")
print()

# Impresión últimas filas
df.tail()
```

# Práctica 12: Agregar una columna

```python
# Impresión de columnas
print('---Columnas de Origen---\n')
print(df.columns)
print()
# Columnas despúes del cambio
df.rename(columns  = {'Unnamed: 0': 'INDEX MIIAD'}, inplace  =  True)
print("---Columnas con cambios---\n")
print(df.columns)
```

# Práctica 13: Normalización de los nombres de la columna

```python
# Normalización de columnas y eliminación de espacios en blanco
df.columns =  df.columns.str.strip().str.lower().str.replace(r"\s+", "_", regex =  True)

print('---Columnas después de normalización')
df.columns
```

# Práctica 14: Errores en los Tipos de Datos

```python
# Impresión de tipos de columna
print('---Tipos de datos por columna (Original)---')
print(df.dtypes)
print()
# Correación de tipos de columna
df['sale_price'] =  pd.to_numeric(df['sale_price'], errors = 'coerce')
df['land_square_feet'] = pd.to_numeric(df['land_square_feet'], errors = 'coerce')
df['gross_square_feet'] =  pd.to_numeric(df['gross_square_feet'], errors =  'coerce')

print('---Tipos de datos con solución de errores---')
print(df.dtypes)
```

# Práctica 15: Generación de Diccionario de Datos

```python
# Generación de diccionario
diccionario_de_datos = pd.DataFrame({'Columna': df.columns, 'Tipo': [str(df[col].dtype) for col in df.columns]})

# Impresión de diccionario
print(diccionario_de_datos)
```
