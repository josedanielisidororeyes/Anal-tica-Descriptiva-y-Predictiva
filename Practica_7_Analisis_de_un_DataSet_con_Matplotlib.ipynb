## Carga de dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
df =  sns.load_dataset('iris')
```

## Verificación de las primeras filas

```python
df.head(2)
```

## Obtener información general sobre los datos

```python
# Impresión de nombres de columnas, tipos de datos, valores núlos, y estadistica descriptiva
df.info()
df.describe()
```

## a) Gráfico de Barras

```python
largo_ancho_x_especie =  df.groupby('species').agg({
    'petal_length': 'mean',
    'petal_width': 'mean',
}).reset_index()


largo_ancho_x_especie.plot(kind = 'bar', x = 'species', figsize = (15, 10))
plt.title('Comparativo del largo y ancho del sépalo por especie')
plt.ylabel('Promedio')
plt.xlabel('Especie')
```

# Histograma

```python
plt.figure(figsize = (5, 5))
df['sepal_length'].hist(bins = 20, color  = 'red')
plt.title('Distribución de los largos de los sépalos')
plt.xlabel('Largo')
plt.ylabel('Frencuencia')
plt.grid(True)
```

# Gráfico de Dispersión

```python
plt.figure(figsize= (10, 10))
for especie, grupo in df.groupby('species'):
  plt.scatter(grupo['petal_length'], grupo['petal_width'], label  = especie)

plt.title('Relación entre el largo y ancho del pétalo por especie')
plt.xlabel('Largo del pétalo')
plt.ylabel('Ancho del pétalo')
plt.legend()
```

# Gráfico de Cajas

```python
fig, axes =  plt.subplots(1, 2, figsize = (12, 6))
df.boxplot(column = 'sepal_width', by  ='species', ax =  axes[0])
axes[0].set_title('Distribución del ancho del sépalo')
axes[0].set_xlabel('Especie')
axes[0].set_ylabel('Medida (cm)')

df.boxplot(column  = 'sepal_length', by  = 'species', ax =  axes[1])
axes[1].set_title('Distriución del largo del sépalo')
axes[1].set_xlabel('Especie')
axes[1].set_ylabel('Medida (cm)')

plt.suptitle('Distribución de las médidas del sépalo')
```

# Gráfico de Líneas

```python
plt.figure(figsize = (12, 12))
df['petal_length'].plot(marker = 'o')
plt.title('Tendencia de la longitud del pétalo')
plt.xlabel('Observación')
plt.ylabel('Medida (cm)')
```
