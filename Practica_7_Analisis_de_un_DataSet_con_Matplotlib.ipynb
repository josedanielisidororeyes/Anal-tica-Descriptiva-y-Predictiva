## Carga de dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
df =  sns.load_dataset('iris')
```

## Verificaci√≥n de las primeras filas

```python
df.head(2)
```

## Obtener informaci√≥n general sobre los datos

```python
# Impresi√≥n de nombres de columnas, tipos de datos, valores n√∫los, y estadistica descriptiva
df.info()
df.describe()
```

## a) Gr√°fico de Barras

```python
largo_ancho_x_especie =  df.groupby('species').agg({
    'petal_length': 'mean',
    'petal_width': 'mean',
}).reset_index()


largo_ancho_x_especie.plot(kind = 'bar', x = 'species', figsize = (15, 10))
plt.title('Comparativo del largo y ancho del s√©palo por especie')
plt.ylabel('Promedio')
plt.xlabel('Especie')
```

# Histograma

```python
plt.figure(figsize = (5, 5))
df['sepal_length'].hist(bins = 20, color  = 'red')
plt.title('Distribuci√≥n de los largos de los s√©palos')
plt.xlabel('Largo')
plt.ylabel('Frencuencia')
plt.grid(True)
```

# Gr√°fico de Dispersi√≥n

```python
plt.figure(figsize= (10, 10))
for especie, grupo in df.groupby('species'):
  plt.scatter(grupo['petal_length'], grupo['petal_width'], label  = especie)

plt.title('Relaci√≥n entre el largo y ancho del p√©talo por especie')
plt.xlabel('Largo del p√©talo')
plt.ylabel('Ancho del p√©talo')
plt.legend()
```

# Gr√°fico de Cajas

```python
fig, axes =  plt.subplots(1, 2, figsize = (12, 6))
df.boxplot(column = 'sepal_width', by  ='species', ax =  axes[0])
axes[0].set_title('Distribuci√≥n del ancho del s√©palo')
axes[0].set_xlabel('Especie')
axes[0].set_ylabel('Medida (cm)')

df.boxplot(column  = 'sepal_length', by  = 'species', ax =  axes[1])
axes[1].set_title('Distriuci√≥n del largo del s√©palo')
axes[1].set_xlabel('Especie')
axes[1].set_ylabel('Medida (cm)')

plt.suptitle('Distribuci√≥n de las m√©didas del s√©palo')
```

# Gr√°fico de L√≠neas

```python
plt.figure(figsize = (12, 12))
df['petal_length'].plot(marker = 'o')
plt.title('Tendencia de la longitud del p√©talo')
plt.xlabel('Observaci√≥n')
plt.ylabel('Medida (cm)')
```
