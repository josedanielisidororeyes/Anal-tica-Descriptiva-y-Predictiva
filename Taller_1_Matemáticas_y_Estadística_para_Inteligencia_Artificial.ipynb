```python
import pandas as pd

df =  pd.read_csv('/content/matriz_dependencias3.csv')
del df['Unnamed: 0']
df.head()
```

```python
A = df.values

# Display the dataframe with indentities
print("DataFrame with row and column identities")
df.head()
```

```python
df.shape
```

```python
# Para hacer un análisis se recomienda tener mas observaciones que variables
# Ya que de esta manera puedo encontrar relaciones entre variables
import numpy as np

# Calcula el rango (estamos buscando combinaciones lineales)
rango  =  np.linalg.matrix_rank(A)
print(f"\n el rango de la mtrix", rango)
print(f"Numero de columns: {A.shape[1]}")
```

```python
from sympy import Matrix

print("\n--- Analizando dependencia con RREF ---")
rref_matrix, pivot_indices =  Matrix(A).rref()

print("\nReduced row Echolon Form (RREF)")
rref_matrix, pivot_indices =  Matrix(A).rref()

print("Pivot column indices (independent)")
print(pivot_indices)

# Obtener nombres de columnas pivote
pivot_columns = [df.columns[i] for i in pivot_indices]
print("Columnas independientes: ", pivot_columns)
```

```python
num_cols  =  A.shape[1]
reff_matrix  =  A
for j in range(num_cols):
  if j not in pivot_indices:
    print(f"\nColumn {j} is dependent on the independent columns:")

    # Build the linear combination string
    combination_str =  ""
    for i, pivot_index in enumerate(pivot_indices):
      coefficient =  reff_matrix[i, j]
      # Only  addd if the coefficient is not zeros
      if coefficient  != 0:
        if combination_str:
          combination_str += " + "
        combination_str += f"({coefficient}) * (C_{pivot_index})"
        print(f"{df.columns[pivot_index]}  (C_{pivot_index})")

    print(f"C_{j} = {combination_str}")
```
